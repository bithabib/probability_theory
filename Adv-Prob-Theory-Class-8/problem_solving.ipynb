```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom, poisson

# Parameters
q = 0.5
lambda_poisson = 2

# Geometric distribution
n_values = np.arange(1, 15)
geometric_pmf = geom.pmf(n_values, q)

# Poisson distribution
poisson_pmf = poisson.pmf(n_values, lambda_poisson)

# Plotting
plt.figure(figsize=(12, 6))

# Geometric Distribution Plot
plt.subplot(1, 2, 1)
plt.stem(n_values, geometric_pmf, use_line_collection=True)
plt.title('Geometric Distribution (q=0.5)')
plt.xlabel('Number of Trials')
plt.ylabel('Probability')

# Poisson Distribution Plot
plt.subplot(1, 2, 2)
plt.stem(n_values, poisson_pmf, use_line_collection=True)
plt.title('Poisson Distribution (λ=2)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')

plt.tight_layout()
plt.show()
```
